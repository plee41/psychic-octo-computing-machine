```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('Desktop/4105/D3.csv')
print(dataset)
dataset.head()
m = len(dataset)
print(m)
```

```python
x0 = np.ones((m, 1))
x1 = dataset.values[:,0]
x2 = dataset.values[:,1]
x3 = dataset.values[:,2]
y = dataset.values[:,3]

print("x0 = ", x0[:10])
print("x1 = ", x1[:10])
print("x2 = ", x2[:10])
print("x3 = ", x3[:10])
print("y = ", y[:10])
```

```python
plt.scatter(x1,y, color='blue', marker='.')
plt.grid() 
plt.xlabel('x1:Input data set') 
plt.ylabel('Y:Output data set') 
plt.title('Scatter plot of training data set 1') 
```

```python
plt.scatter(x2,y, color='red', marker='.')
plt.grid() 
plt.xlabel('x2:Input data set') 
plt.ylabel('Y:Output data set') 
plt.title('Scatter plot of training data set 2') 
```

```python
plt.scatter(x3,y, color='green', marker='.')
plt.grid() 
plt.xlabel('x3:Input data set') 
plt.ylabel('Y:Output data set') 
plt.title('Scatter plot of training data set 3') 
```

```python
x1 = x1.reshape(m,1)
x2 = x2.reshape(m,1)
x3 = x3.reshape(m,1)
print("x1 = ", x1[:10])
print("x2 = ", x2[:10])
print("x3 = ", x3[:10])
```

```python
X_1 = np.hstack((x0, x1))
X_2 = np.hstack((x0, x2))
X_3 = np.hstack((x0, x3))
print("x1 = ", X_1[:10])
print("x2 = ", X_2[:10])
print("x3 = ", X_3[:10])
```
